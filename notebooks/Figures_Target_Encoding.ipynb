## Beta encoder

```python
N =100
n = 75
prior_mean = 0.5
n_mins = range(0, 151)
beta_vals = []
for N_min in n_mins:
    N_prior = np.maximum(N_min-N, 0)
    prior_mean*N_prior

    alpha = (1-prior_mean)* N_prior + n
    beta = (1-prior_mean) * N_prior + N-n
    beta_vals += [alpha/(alpha+beta)]
```

```python
# plt.axvline(75, color="black", linestyle="--", label="n")


plt.ylim(0.45, 0.8)
plt.plot(pd.Series(data = vals_m, index = range_m), label = "m_Estimate")
# plt.plot((vals_m * check_random_state(42).normal(1., 0.05, len(range_m))), label="M_Estimate_reg")
plt.plot(pd.Series(data = beta_vals, index = n_mins), label = "Beta")
plt.axvline(100, color="grey", linestyle="--", label="Ni")
plt.xlabel("m and N_min", fontsize=12)
plt.ylabel("Encoding", fontsize=12)
plt.legend(fontsize=10)
plt.suptitle('M_Estimate and Beta Encoding', fontsize=14)
```
